### No PU Samples

```python
f = uproot.open("files/125X_mcRun3_2022_realistic_v3_nopu_ckf.root")
tree = f[f.keys()[0]]
data_nopu_125_ckf = pd.DataFrame(tree.arrays(),columns=tree.keys())
f = uproot.open("files/125X_mcRun3_2022_realistic_v3_nopu_mkfit.root")
tree = f[f.keys()[0]]
data_nopu_125_mkf = pd.DataFrame(tree.arrays(),columns=tree.keys())

f = uproot.open("files/124X_mcRun3_2022_realistic_v7_nopu_ckf.root")
tree = f[f.keys()[0]]
data_nopu_124_ckf = pd.DataFrame(tree.arrays(),columns=tree.keys())
f = uproot.open("files/124X_mcRun3_2022_realistic_v7_nopu_mkfit.root")
tree = f[f.keys()[0]]
data_nopu_124_mkf = pd.DataFrame(tree.arrays(),columns=tree.keys())
```

```python
data_nopu_125_ckf.head()
```

```python
data_nopu_125_ckf["XI_DX"] = data_nopu_125_ckf["XI_VtxX"]-data_nopu_125_ckf["PV_becos_XX"]
data_nopu_125_ckf["XI_DZ"] = data_nopu_125_ckf["XI_VtxZ"]-data_nopu_125_ckf["PV_becos_ZZ"]
data_nopu_125_ckf["XI_DY"] = data_nopu_125_ckf["XI_VtxY"]-data_nopu_125_ckf["PV_becos_YY"]
```

```python
data_nopu_125_mkf["XI_DX"] = data_nopu_125_mkf["XI_VtxX"]-data_nopu_125_mkf["PV_becos_XX"]
data_nopu_125_mkf["XI_DZ"] = data_nopu_125_mkf["XI_VtxZ"]-data_nopu_125_mkf["PV_becos_ZZ"]
data_nopu_125_mkf["XI_DY"] = data_nopu_125_mkf["XI_VtxY"]-data_nopu_125_mkf["PV_becos_YY"]
```

```python
data_nopu_125_ckf["XI_R"] = (data_nopu_125_ckf["XI_DY"].values**2 +data_nopu_125_ckf["XI_DX"].values**2 +data_nopu_125_ckf["XI_DZ"].values**2)**0.5
```

```python
data_nopu_125_mkf["XI_R"] = (data_nopu_125_mkf["XI_DY"].values**2 +data_nopu_125_mkf["XI_DX"].values**2 +data_nopu_125_mkf["XI_DZ"].values**2)**0.5
```

```python
xi_r_125_mkf = np.array([f for a in data_nopu_125_mkf["XI_R"] for f in a])
xi_r_125_ckf = np.array([f for a in data_nopu_125_ckf["XI_R"] for f in a])
```

```python
B = 100
R = (0.0,50.0)
BW = (R[1]-R[0])/B * 1000

n_mkfit = data_nopu_125_mkf["nCand"].mean()
n_ckf = data_nopu_125_ckf["nCand"].mean()

plt.hist(xi_r_125_mkf,range=R,bins=B,color="blue",label="New MkFit - N(Xi) = %.2f"%(n_mkfit),log=True);
plt.hist(xi_r_125_ckf,range=R,bins=B,histtype="step",color="red",label="CKF - N(Xi) = %.2f"%(n_ckf),log=True);
plt.legend(fontsize = 16)
plt.xlabel("$M(Xi)$",fontsize=14)
plt.ylabel("Entries / %.3f MeV"%BW,fontsize=14)

plt.title("New 12_5_0 samples - Xi R (w.r.t. PV)",fontsize=20)
```

```python
B = 100
R = (0.0,50.0)
BW = (R[1]-R[0])/B * 1000

n_mkfit = data_nopu_125_mkf["nCand"].mean()
n_ckf = data_nopu_125_ckf["nCand"].mean()

plt.hist(xi_r_125_mkf,range=R,bins=B,color="blue",label="New MkFit - N(Xi) = %.2f"%(n_mkfit),log=True);
plt.hist(xi_r_125_ckf,range=R,bins=B,histtype="step",color="red",label="CKF - N(Xi) = %.2f"%(n_ckf),log=True);
plt.legend(fontsize = 16)
plt.xlabel("$M(Xi)$",fontsize=14)
plt.ylabel("Entries / %.3f MeV"%BW,fontsize=14)

plt.title("New 12_5_0 samples - Xi R (w.r.t. PV)",fontsize=20)
```

```python
xi_mass_125_mkf = np.array([f for a in data_nopu_125_mkf["XI_mass"] for f in a])
xi_mass_125_ckf = np.array([f for a in data_nopu_125_ckf["XI_mass"] for f in a])
```

```python
xi_mass_124_mkf = np.array([f for a in data_nopu_124_mkf["XI_mass"] for f in a])
xi_mass_124_ckf = np.array([f for a in data_nopu_124_ckf["XI_mass"] for f in a])
```

```python
B = 100
R = (1.25,1.37)
BW = (R[1]-R[0])/B * 1000
```

```python
sig_max = 1.33
sig_min = 1.315
```

```python
xi_mass_125_mkf[(xi_mass_125_mkf<=sig_max) & (xi_mass_125_mkf>=sig_min)]
```

```python
xi_mass_125_ckf[(xi_mass_125_ckf<=sig_max) & (xi_mass_125_ckf>=sig_min)]
```

```python
n_mkfit = data_nopu_125_mkf["nCand"].sum()
n_ckf = data_nopu_125_ckf["nCand"].sum()

plt.hist(xi_mass_125_mkf,range=R,bins=B,color="blue",label="New MkFit - N(Xi) = %d"%(n_mkfit));
plt.hist(xi_mass_125_ckf,range=R,bins=B,histtype="step",color="red",label="CKF - N(Xi) = %d"%(n_ckf));
plt.legend(fontsize = 16)
plt.xlabel("$M(Xi)$",fontsize=14)
plt.ylabel("Entries / %.3f MeV"%BW,fontsize=14)
plt.plot([sig_min,sig_min],[0,400],"--",color="green")
plt.plot([sig_max,sig_max],[0,400],"--",color="green")

sig_size_mkf = len(xi_mass_125_mkf[(xi_mass_125_mkf<=sig_max) & (xi_mass_125_mkf>=sig_min)])
sig_size_ckf = len(xi_mass_125_ckf[(xi_mass_125_ckf<=sig_max) & (xi_mass_125_ckf>=sig_min)])

plt.text(1.25,100,"N(mkfit)/N(ckf) = %.2f [all over the spectrum]"%(n_mkfit/float(n_ckf)))
plt.text(1.25,80,"N(mkfit)/N(ckf) = %.2f [in signal region]"%(sig_size_mkf/float(sig_size_ckf)))

plt.title("New 12_5_0 samples",fontsize=20)

```

```python
n_mkfit = data_nopu_124_mkf["nCand"].sum()
n_ckf = data_nopu_124_ckf["nCand"].sum()

plt.hist(xi_mass_124_mkf,range=R,bins=B,color="blue",label="New MkFit - N(Xi) = %d"%(n_mkfit));
plt.hist(xi_mass_124_ckf,range=R,bins=B,histtype="step",color="red",label="CKF - N(Xi) = %d"%(n_ckf));
plt.legend(fontsize = 16)
plt.xlabel("$M(Xi)$",fontsize=14)
plt.ylabel("Entries / %.3f MeV"%BW,fontsize=14)
plt.plot([sig_min,sig_min],[0,100],"--",color="green")
plt.plot([sig_max,sig_max],[0,100],"--",color="green")

sig_size_mkf = len(xi_mass_124_mkf[(xi_mass_124_mkf<=sig_max) & (xi_mass_124_mkf>=sig_min)])
sig_size_ckf = len(xi_mass_124_ckf[(xi_mass_124_ckf<=sig_max) & (xi_mass_124_ckf>=sig_min)])

plt.text(1.25,40,"N(mkfit)/N(ckf) = %.2f [all over the spectrum]"%(n_mkfit/float(n_ckf)))
plt.text(1.25,30,"N(mkfit)/N(ckf) = %.2f [in signal region]"%(sig_size_mkf/float(sig_size_ckf)))

plt.title("Old 12_4_1 samples",fontsize=20)
```
